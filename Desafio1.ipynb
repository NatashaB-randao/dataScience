# Importe o arquivo alunos.csv e armazene seu conteúdo em um DataFrame Pandas.

 **Para importar o arquivo em questão e salvar ele em um DataFrame, podemos importar a biblioteca Pandas e utilizar a função read_csv() passando a url onde esse arquivo está localizado:**

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
```

## Calcule algumas estatísticas descritivas básicas dos dados do DataFrame (média, desvio padrão, etc). Dica: pesquise pelo método describe.

**Como a própria dica já contou, podemos utilizar um método do Pandas chamado describe para calcularmos algumas estatísticas descritivas básicas dos dados do DataFrame:**

```python
dados.describe()
```

**O método describe() é uma das funções mais úteis do Pandas para a análise exploratória de dados. Essa função é usada para calcular algumas estatísticas descritivas básicas dos dados em um DataFrame ou em uma coluna específica de um DataFrame.**
