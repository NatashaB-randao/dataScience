```python
import pandas as pd
```

```python
dados = pd.read_csv('alunos.csv')
```

## Verifique se a base de dados possui dados nulos e, caso tenha, realize o tratamento desses dados nulos da forma que achar mais coerente com a situação.

**Podemos começar verificando a existência de dados nulos:**

```python
dados.isnull().sum()
```

**E com o resultado, vamos notar que apenas a coluna "Notas" possui dados nulos. Nesse caso, podemos pensar que os alunos que estão sem nota muito provavelmente estavam com 0. Sendo assim, podemos substituir os valores nulos por 0:**

```python
dados = dados.fillna(0)
dados
```

## Os alunos "Alice" e "Carlos", não fazem mais parte da turma. Sendo assim, remova-os da base de dados.

**Para removermos os alunos "Alice" e "Carlos" da nossa base de dados, precisamos pegar os índices das linhas onde os nomes deles estão localizados e utilizar o método drop para remover as respectivas linhas:**

```python
# salvando os índices das linhas a serem removidas em uma variável
alunos_a_remover = dados.query('Nome == "Alice" | Nome == "Carlos"').index
```

```python
# removendo as linhas referentes aos alunos "Alice" e "Carlos"
dados.drop(alunos_a_remover, axis=0, inplace=True)
```

```python
dados
```

## Aplique um filtro que selecione apenas os alunos que foram aprovados.

**Nesse caso, nosso primeiro passo é criar uma variável onde nós realizamos a seleção apenas dos alunos aprovados:**

```python
selecao = dados['Aprovado'] == True
```

**Feito isso, podemos passar essa seleção para nosso DataFrame:**

```python
dados[selecao]
```

##  Salve o DataFrame que possui apenas os alunos aprovados em um arquivo csv chamado "alunos_aprovados.csv".

**Para salvarmos o DataFrame apenas com os alunos aprovados, nós podemos começar salvando o DataFrame resultante do desafio 3 em uma variável:**

```python
alunos_aprovados = dados[selecao]
```

**Em seguida, podemos utilizar o método to_csv para salvá-lo no formato csv:**

```python
alunos_aprovados.to_csv('alunos_aprovados.csv', index=False)
```

## Extra: Ao conferir as notas dos alunos aprovados, notamos que algumas notas estavam incorretas. As alunas que tiraram nota 7.0, na verdade, tinham um ponto extra que não foi contabilizado. Sendo assim, substitua as notas 7.0 da base de dados por 8.0. Dica: pesquise pelo método replace.


**O método replace é utilizado para substituir um determinado valor por outro. No caso da nossa base de dados, nós queremos substituir os valores 7.0 por 8.0. Para isso, nós podemos utilizar o seguinte trecho de código:**

```python
alunos_aprovados.replace(7.0, 8.0)
```

**E para salvarmos essa alteração de forma definitiva, podemos atribuí-la a nossa variável:**

```python
alunos_aprovados = alunos_aprovados.replace(7.0, 8.0)
alunos_aprovados
```
