```python
import pandas as pd
```

```python
dados = pd.read_csv('alunos.csv')
```

## Os alunos participaram de uma atividade extracurricular e ganharam pontos extras. Esses pontos extras correspondem a 40% da nota atual de cada um deles. Com base nisso, crie uma coluna chamada "Pontos_extras" que contenha os pontos extras de cada aluno, ou seja, 40% da nota atual deles.


**Para criar a coluna de "Pontos_extras", podemos utilizar o método apply para aplicar uma função na coluna "Notas". Essa função deve apenas multiplicar 0.4 pelo valor da nota de cada aluno e colocar o resultado na nossa nova coluna:**

```python
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x*0.4)
dados
```

## Crie mais uma coluna, chamada "Notas_finais" que possua as notas de cada aluno somada com os pontos extras.

**A coluna "Notas_finais" vai ser a soma das colunas "Notas" e "Pontos_extras":**

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados
```

## Como houve uma pontuação extra, alguns alunos que não tinham sido aprovados antes podem ter sido aprovados agora. Com base nisso, crie uma coluna chamada "Aprovado_final" com os seguintes valores:

True: caso o aluno esteja aprovado (nota final deve ser maior ou igual a 6);
False: caso o aluno esteja reprovado (nota final deve ser menor que 6).


**Para a criação da coluna "Aprovado_final", nós podemos utilizar o método apply para aplicar uma função na coluna "Notas_finais". Essa função deve colocar True, caso a nota final seja maior ou igual a 6, e, caso contrário, ela deve colocar False:**

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x: True if x >= 6 else False)
dados
```

## Faça uma seleção e verifique quais alunos não tinham sido aprovados anteriormente, mas foram aprovados após a soma dos pontos extras.

**Essa seleção pode ser realizada utilizando o método query:**

```python
dados.query('Aprovado == False & Aprovado_final == True')
```

**Ou também pode ser feita de forma mais manual:**

```python
selecao = (dados['Aprovado'] == False) & (dados['Aprovado_final'] == True)
dados[selecao]
```
